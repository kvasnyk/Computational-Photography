```python
import numpy
from scipy import ndimage
```

```python
bearImg = ndimage.imread('./data/bear.png')
bearMaskImg = ndimage.imread('./data/bear_mask.png')
waterpoolImg = ndimage.imread('./data/waterpool.png')
```

```python
def naiveComposite(bg, fg, mask, y, x):
    fgWidth = fg.shape[1]
    fgHeight = fg.shape[0]
    
    maskNormal = mask / 255
    
    fgMasked = fg * maskNormal
    
    bgSlice = bg[y : y + fgHeight, x : x + fgWidth, :]
    bgSliceMasked = bgSlice * (1 - maskNormal)
    
    result = numpy.copy(bg)
    result[y : y + fgHeight, x : x + fgWidth, :] = fgMasked + bgSliceMasked
    
    return result.astype(uint8)
```

```python
result1 = naiveComposite(waterpoolImg, bearImg, bearMaskImg, 10, 10)
result2 = naiveComposite(waterpoolImg, bearImg, bearMaskImg, 100, 10)
result3 = naiveComposite(waterpoolImg, bearImg, bearMaskImg, 50, 30)
```
