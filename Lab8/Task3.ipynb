```python
import numpy
import scipy
from scipy import signal
from scipy import ndimage
from scipy import misc
```

```python
rampImg = ndimage.imread('./data/ramp.png') / 255.0
flatImg = ndimage.imread('./data/flat.png') / 255.0
smallMaskImg = ndimage.imread('./data/small_mask.png') / 255.0
```

```python
def laplacian(image):
    mat = np.array([[0., -1., 0.], [-1., 4., -1.], [0., -1., 0.]])
    
    r = scipy.ndimage.filters.convolve(image[:, :, 0], mat)
    g = scipy.ndimage.filters.convolve(image[:, :, 1], mat)
    b = scipy.ndimage.filters.convolve(image[:, :, 2], mat)
    
    result = numpy.copy(image)
    result[:, :, 0] = r
    result[:, :, 1] = g
    result[:, :, 2] = b
    
    return scipy.ndimage.filters.laplace(image)
    
    return result
```

```python
def dotIm(im1, im2):
    return sum(im1 * im2)
```

```python
def poissonCG(bg, fg, mask, niter):
    mask = mask[:, :, 0:3]
    
    b = laplacian(fg)
    
    x = bg * (1.0 - mask)
    r = (b - laplacian(x)) * mask
    d = (b - laplacian(x)) * mask
    
    for i in range(niter):
        alpha = dotIm(r, r) / dotIm(d, laplacian(d))
        xNew = x + alpha * d
        rNew = (r - alpha * laplacian(d)) * mask
        beta = dotIm(rNew, rNew) / dotIm(r, r)
        dNew = (rNew + beta * d) * mask
        
        x = xNew
        r = rNew
        d = dNew
    
    return x
```

```python
poissonResult1 = poissonCG(rampImg, flatImg, smallMaskImg, 0)
poissonResult2 = poissonCG(rampImg, flatImg, smallMaskImg, 5)
poissonResult3 = poissonCG(rampImg, flatImg, smallMaskImg, 10)
poissonResult4 = poissonCG(rampImg, flatImg, smallMaskImg, 15)
```

```python
def poissonCGComposite(bg, fg, mask, y, x, niter):
    fgWidth = fg.shape[1]
    fgHeight = fg.shape[0]
    
    bgSlice = bg[y : y + fgHeight, x : x + fgWidth, :]
    
    result = numpy.copy(bg)
    resultSlice = poissonCG(bgSlice, fg, mask, niter)
    result[y : y + fgHeight, x : x + fgWidth, :] = resultSlice
    
    return result
```

```python
bearImg = ndimage.imread('./data/bear.png') / 255.0
bearMaskImg = ndimage.imread('./data/bear_mask.png') / 255.0
waterpoolImg = ndimage.imread('./data/waterpool.png') / 255.0
```

```python
result1 = poissonCGComposite(waterpoolImg, bearImg, bearMaskImg, 10, 10, 200)
result2 = poissonCGComposite(waterpoolImg, bearImg, bearMaskImg, 100, 10, 200)
result3 = poissonCGComposite(waterpoolImg, bearImg, bearMaskImg, 50, 30, 200)
```

```python
pyplot.subplot(1, 3, 1)
pyplot.axis('off')
pyplot.imshow(result1)

pyplot.subplot(1, 3, 2)
pyplot.axis('off')
pyplot.imshow(result2)

pyplot.subplot(1, 3, 3)
pyplot.axis('off')
pyplot.imshow(result3)
```
