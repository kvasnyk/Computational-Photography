```python
import numpy
from scipy import ndimage
```

```python
probesPaths = [
    './probes/grace_probe.npy',
    './probes/rnl_probe.npy',
    './probes/galileo_probe.npy',
    './probes/uffizi_probe.npy'
]

probes = map(numpy.load, probesPaths)
```

```python
thetas = [0, 45, 90, 135, 180, 225, 270, 315]
phis = [-60, -30, 0, 30, 60]
```

```python
teapots = []
coords = []

for theta in thetas:
    for phi in phis:        
        teapotPath = './relighting/teapot_' + str(theta) + '_' + str(phi) + '.png'
        teapot = ndimage.imread(teapotPath) / 255.0
        
        teapots.append(teapot)
        coords.append([theta, phi])
        
coords = numpy.array(coords)
```

```python
dx = numpy.cos(coords[:, 1]) * numpy.sin(coords[:, 0])
dy = numpy.cos(coords[:, 0])
dz = - numpy.sin(coords[:, 1]) * numpy.sin(coords[:, 0])
r = (1 / numpy.pi) * numpy.arccos(dz) / numpy.sqrt(dx ** 2 + dy ** 2)
u = (dx * r + 1.) / 2.
v = (dy * r + 1.) / 2.
```

```python
def computeProbeColours(probeIndex):
    up = u * probes[probeIndex].shape[0]
    vp = v * probes[probeIndex].shape[1]
    
    return map(lambda i: probes[probeIndex][up[i]][vp[i]], range(len(coords)))
```

```python
probesColours = map(lambda i: computeProbeColours(i), range(len(probes)))
```

```python
def colourImage(probeIndex, coordsIndex):
    image = teapots[coordsIndex]
    return image * probesColours[probeIndex][coordsIndex]
```

```python
def relight(probeIndex):
    colouredImages = map(lambda i: colourImage(probeIndex, i), range(len(coords)))
    return numpy.sum(colouredImages, axis = 0)
```

```python
results = map(lambda i: relight(i), range(len(probes)))
```
